```python
# Obejct Oriented Methode
fig = plt.figure()
# axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes=fig.add_axes([0.5,0.5,0.8,0.8])
axes.plot(x,y)
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(y,x)
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
axes.plot(x,y,'r')
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X axes')
axes.set_ylabel('Y axes')
axes.set_title('Title')
axes.plot(x,y)
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_xlabel('X1 Axis')
axes1.set_ylabel('Y1 Axis')
axes1.set_title('First figure')
axes1.plot(x,y)
axes2=fig.add_axes([0.2,0.6,0.3,0.2])
axes2.set_xlabel('X2 Axis')
axes2.set_ylabel('Y2 Axis')
axes2.set_title('Second figure')
axes2.plot(y,x)

plt.show()
```
