```python
race_year=race_result['year']
```

```python
ra_year=pd.DataFrame(race_year)
```

```python
driver_name=race_result['name']
```

```python
dr_name=pd.DataFrame(driver_name)
```

```python
driver_nationality=race_result['nationality']
```

```python
dr_na=pd.DataFrame(driver_nationality)
```

```python
total_points=race_result['points']
```

```python
to_po=pd.DataFrame(total_points)
```

```python
drivers_stand=ra_year.join(dr_name).join(dr_na).join(to_po)
```

```python
drivers_stand.rename(columns={'year':'race_year','name':'driver_name','nationality':'driver_nationality','points':'total_points'},inplace=True)
```

```python
drivers_stand
```

```python
drivers_stand.to_csv('drivers_standing.csv')
```
